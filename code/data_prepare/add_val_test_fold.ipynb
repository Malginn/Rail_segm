```python
path_train_imgs = '../data/Railsem19.v7i.yolov8/train/images'
path_train_labels = '../data/Railsem19.v7i.yolov8/train/labels'
path_valid_imgs = '../data/Railsem19.v7i.yolov8/valid/images'
path_valid_labels = '../data/Railsem19.v7i.yolov8/valid/labels'
path_test_imgs = '../data/Railsem19.v7i.yolov8/test/images'
path_test_labels = '../data/Railsem19.v7i.yolov8/test/labels'



onlyfiles = [f for f in listdir(path_train_imgs) if isfile(join(path_train_imgs, f))]
onlyfiles

# shutil.move("path/to/current/file.foo", "path/to/new/destination/for/file.foo")

def move_objs(dir_from):    #, dir_to
    onlyfiles = [f for f in listdir(dir_from) if isfile(join(dir_from, f))]
    random.shuffle(onlyfiles)
    
    to_move_imgs = onlyfiles[:500]

    for i in to_move_imgs:
        shutil.move(f"../data/Railsem19.v7i.yolov8/train/images/{i}", "../data/Railsem19.v7i.yolov8/test/images")
        shutil.move(f"../data/Railsem19.v7i.yolov8/train/labels/{i[:-4] + '.txt'}", "../data/Railsem19.v7i.yolov8/test/labels")


move_objs(path_train_imgs)
```
